```python
import numpy as np
import matplotlib.pyplot as plt
import powerlaw
```

```python
path = 'Training_Data_Master/'
syslen = []
syscall = []

for i in range(1, 834):
    if (i > 0 and i < 10):
        with open(path + 'UTD-000{}.txt'.format(i)) as f:
            lines = f.readlines()
            for item in lines:
                syscall.extend(item.split())
            syslen.append(len(syscall))
            syscall = []
    
    if (i >= 10 and i < 100):
        with open(path + 'UTD-00{}.txt'.format(i)) as f:
            lines = f.readlines()
            for item in lines:
                syscall.extend(item.split())
            syslen.append(len(syscall))
            syscall = []
    
    if (i >= 100 and i < 1000):
        with open(path + 'UTD-0{}.txt'.format(i)) as f:
            lines = f.readlines()
            for item in lines:
                syscall.extend(item.split())
            syslen.append(len(syscall))
            syscall = []

print("system call length:\n", syslen, len(syslen))

```

```python
# calculator frequency
syslen.sort()

# plot raw data
Y = np.array(syslen)
plt.title("Train Raw Data")
plt.hist(Y, label='train data')
plt.ylabel('system call length')
plt.xlabel('count')


def DataVisualization():
    # 繪製 log-log plot
    pass

# find powerlaw
results = powerlaw.Fit(Y)
print("alpha:", results.power_law.alpha)
print("Xmin:", results.power_law.xmin)
R, p = results.distribution_compare('power_law', 'lognormal')

```
